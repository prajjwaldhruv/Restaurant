<h3>ADDING EXTRA COLUMNS TO THE rating_final TABLE</h3>

```python
students=[]
professionals=[]
unemployeds=[]
workingclass=[]
unknowns=[]
k=0
for i in userprofile['activity']:
    if i=='student':
        students.append(userprofile['userID'][k])
    elif i=='professional':
        professionals.append(userprofile['userID'][k])
    elif i=='unemployed':
        unemployeds.append(userprofile['userID'][k])
    elif i=='working-class':
        workingclass.append(userprofile['userID'][k])
    else:
        unknowns.append(userprofile['userID'][k])
    k+=1

f=[]
for i in rating_final['userID']:
    if i in students:
        f.append('Student')
    elif i in professionals:
        f.append('Professional')
    elif i in unemployeds:
        f.append('Unemployed')
    elif i in workingclass:
        f.append('Working-Class')
    else:
        f.append('Unknown')

rating_final['activity']=f
rating_final.head()
```

<h4>ADDING A COLUMN THAT SHOWS MARRIED STATUS OF ALL THE USERS</h4>

```python
single=[]
married=[]
widow=[]
unknown=[]
k=0
for i in userprofile['marital_status']:
    if i=='single':
        single.append(userprofile['userID'][k])
    elif i=='married':
        married.append(userprofile['userID'][k])
    elif i=='widow':
        widow.append(userprofile['userID'][k])
    else:
        unknown.append(userprofile['userID'][k])
    k+=1

f=[]
for i in rating_final['userID']:
    if i in single:
        f.append('Single')
    elif i in married:
        f.append('Married')
    elif i in widow:
        f.append('Widow')
    else:
        f.append('Unknown')

rating_final['married_status']=f
rating_final.head()
```

<h4>ADDING COLUMNS SHOWING WHETHER THE USER IS SMOKER OR NOT AND IF THE RESTAURANT HAS SMOKING ALLOWEWD</h4>

```python
nonsmoker=[]
smoker=[]
k=0
for i in userprofile['smoker']:
    if i=='true':
        smoker.append(userprofile['userID'][k])
    else:
        nonsmoker.append(userprofile['userID'][k])
    k+=1

f=[]
for i in rating_final['userID']:
    if i in smoker:
        f.append('Smoker')
    else:
        f.append('nonSmoker')

rating_final['smoker']=f

none=[]
onlyatbar=[]
permitted=[]
section=[]
notpermitted=[]
l=0
for i in geoplaces2['smoking_area']:
    if i=='none':
        none.append(geoplaces2['placeID'][l])
    elif i=='only at bar':
        onlyatbar.append(geoplaces2['placeID'][l])
    elif i=='permitted':
        permitted.append(geoplaces2['placeID'][l])
    elif i=='section':
        section.append(geoplaces2['placeID'][l])
    else:
        notpermitted.append(geoplaces2['placeID'][l])
    l+=1

g=[]
for i in rating_final['placeID']:
    if i in none:
        g.append('None')
    elif i in onlyatbar:
        g.append('Only at bar')
    elif i in permitted:
        g.append('Permitted')
    elif i in section:
        g.append('Section')
    elif i in notpermitted:
        g.append('Not permitted')
    else:
        g.append('')
        
rating_final['res_smokingarea']=g

rating_final.head()
```

```python
noalcohol=[]
winebeer=[]
fullbar=[]
k=0
for i in geoplaces2['alcohol']:
    if i=='No_Alcohol_Served':
        noalcohol.append(geoplaces2['placeID'][k])
    elif i=='Wine-Beer':
        winebeer.append(geoplaces2['placeID'][k])
    else:
        fullbar.append(geoplaces2['placeID'][k])
    k+=1

f=[]
for i in rating_final['placeID']:
    if i in noalcohol:
        f.append('No_Alcohol_Served')
    elif i in winebeer:
        f.append('Wine-Beer')
    else:
        f.append('Full_Bar')

rating_final['alcohol']=f

abstemious=[]
casualdrinker=[]
socialdrinker=[]
l=0
for i in userprofile['drink_level']:
    if i=='abstemious':
        abstemious.append(userprofile['userID'][l])
    elif i=='casual drinker':
        casualdrinker.append(userprofile['userID'][l])
    else:
        socialdrinker.append(userprofile['userID'][l])
    l+=1

g=[]
for i in rating_final['userID']:
    if i in abstemious:
        g.append('Abstemious')
    elif i in casualdrinker:
        g.append('Casual Drinker')
    else:
        g.append('Social Drinker')
        
rating_final['drink_level']=g

rating_final.head()
```

```python
low=[]
medium=[]
high=[]
unknown=[]
k=0
for i in userprofile['budget']:
    if i=='low':
        low.append(userprofile['userID'][k])
    elif i=='medium':
        medium.append(userprofile['userID'][k])
    elif i=='high':
        high.append(userprofile['userID'][k])
    else:
        unknown.append(userprofile['userID'][k])
    k+=1

f=[]
for i in rating_final['userID']:
    if i in low:
        f.append('low')
    elif i in medium:
        f.append('medium')
    elif i in high:
        f.append('high')
    else:
        f.append('unknown')
rating_final['budget']=f

rating_final.head()
```

<h5>In the following graph I tried to see how smoker and non-smoker users rate restaurants where smoking is not allowed differently</h5>

```python
a=0
b=0
c=0
d=0
e=0
f=0
g=0
k=0
l=0
q=0
for i in rating_final['smoker']:
    if rating_final['res_smokingarea'][q] == 'Not permitted' or rating_final['res_smokingarea'][q] == 'None':
        if i=='Smoker':
            if rating_final['service_rating'][q] == 0:
                a+=1
            elif rating_final['service_rating'][q] == 1:
                b+=1
            else:
                c+=1
            k+=1
        else:
            if rating_final['service_rating'][q] == 0:
                d+=1
            elif rating_final['service_rating'][q] == 1:
                e+=1
            else:
                f+=1
            l+=1
    else:
        g+=1
    q+=1
            
smokers=[]
nonsmokers=[]
smokers.append(a*100/k)
smokers.append(b*100/k)
smokers.append(c*100/k)
nonsmokers.append(d*100/l)
nonsmokers.append(e*100/l)
nonsmokers.append(f*100/l)

barWidth = 0.25
r1 = np.arange(len(smokers))
r2 = [x + barWidth for x in r1]

plt.figure(figsize=(16,8))
plt.bar(r1, smokers, color='#F4D59A', width=barWidth, edgecolor='white', label='Smoker')
plt.bar(r2, nonsmokers, color='#DE8626', width=barWidth, edgecolor='white', label='Non-smoker')
 
plt.xlabel('Ratings', fontweight='bold')
plt.ylabel('Percentage of users', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(smokers))], ['0','1','2'])
 
plt.legend()
plt.show()
```
